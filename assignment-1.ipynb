```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 2
b = 3
x = np.arange(-10,10,0.1)
y = a*x+b

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f"y = {a}x + {b}")
plt.show()
```

```python
a = 2
b = 3
x = np.linspace(-10,10,100)
y = a*np.sin(x)+b

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f"y = {a}sin(x) + {b}")
plt.show()
```

```python
a = 2
b = 3
x = np.linspace(-10,10,100)
y = a*np.exp(x)+b

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f"y = {a}e^x + {b}")
plt.show()
```

```python
a = 2
b = 3
x = np.linspace(-10, 10, 100)
y = a * (x**2) + b

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"y = {a}x^2 + {b}")
plt.show()
```
